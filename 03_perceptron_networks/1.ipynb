```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import clear_output
import time
```

```python
SEED = 211
NUM_POINTS = 100

# Decision line we're trying to predict
t_w1, t_w2, b = 0.49671415, -0.1382643, 0.64768854
```

```python
#### Generate random test data for the perceptron to train one ####

def generate_linearly_separable_data(NUM_POINTS, w1, w2, b, seed=None):
    np.random.seed(seed)

    # Generate random points
    points = np.random.randn(NUM_POINTS, 2)

    # Compute the labels for the points based on their position relative to the line
    labels = np.sign(w1 * points[:, 0] + w2 * points[:, 1] + b)

    return points, labels


def plot_linearly_separable_data(points, labels, w1, w2, b):
    plt.scatter(points[labels == 1, 0],
                points[labels == 1, 1], c='blue', label='Class 1')
    plt.scatter(points[labels == -1, 0],
                points[labels == -1, 1], c='red', label='Class -1')

    # Plot the separating line
    x1 = np.linspace(points[:, 0].min() - 1, points[:, 0].max() + 1, 100)
    x2 = -(w1 * x1 + b) / w2
    plt.plot(x1, x2, 'g--', label='Separating Line')

    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.legend()
    plt.show()


points, labels = generate_linearly_separable_data(
    NUM_POINTS, t_w1, t_w2, b, seed=SEED)
plot_linearly_separable_data(points, labels, t_w1, t_w2, b)
```
