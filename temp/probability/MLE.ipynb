# Estimate $X \sim Pareto(\alpha)$ :

- <ins>Pareto PDF</ins>:
  $\newline f(x) = \frac{\alpha }{x^{\alpha+1}} $

- <ins>Likelihood</ins>:
$\newline L(\alpha) = \prod^n{ \frac{\alpha}{x_i^{\alpha + 1}}} $

- <ins>Log-Likelihood</ins>:
$\newline LL(\alpha) = \sum^n{ log \frac{\alpha}{x_i^{\alpha + 1}}} =  \sum^n{ log(\alpha) - ({\alpha + 1})log({x_i})} $

- <ins>Gradient of Log-Likelihood</ins>:
 $\newline \frac{\delta LL(\alpha)}{\delta \alpha} = \frac{n}{\alpha}\sum^n{ log({x_i})} $

- <ins>Estimate Alpha</ins>:
$\newline \underset{\alpha}{\operatorname{argmax}}(LL(\alpha)) = \frac{n}{\alpha}\sum^n{ log({x_i})} $



```python
import math

def estimate_alpha(samples):
    log_sum = 0
    for x_i in samples:
        log_sum += math.log(x_i)
    
    n = len(samples)

    return n/log_sum

print(estimate_alpha(samples))
```
